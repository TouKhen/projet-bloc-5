```python
# Python
import pandas as pd

user_id = "AHXNZ75VJUMZ34ASD2BFHTPRD6BA"

review_df = review_df.copy()
review_df["timestamp"] = pd.to_datetime(review_df["timestamp"], errors="coerce")

user_reviews = (
    review_df.loc[review_df["user_id"] == user_id]
    .dropna(subset=["timestamp", "parent_asin"])
    .sort_values("timestamp", ascending=False)
)

last_6 = user_reviews.head(6).reset_index(drop=True)

held_out_latest = last_6.iloc[0].copy()   # keep in memory
context_5 = last_6.iloc[1:6].copy()       # use for prediction

true_next_item = held_out_latest["parent_asin"]
context_items = list(context_5.sort_values("timestamp")["parent_asin"])  # oldest -> newest

true_next_item, context_items
```

```python
# Python
import numpy as np

df = review_df.dropna(subset=["user_id", "timestamp", "parent_asin"]).copy()
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")
df = df.dropna(subset=["timestamp"]).sort_values(["user_id", "timestamp"])

df["next_parent_asin"] = df.groupby("user_id")["parent_asin"].shift(-1)

pairs = df.dropna(subset=["next_parent_asin"])[["parent_asin", "next_parent_asin"]]
pairs.columns = ["prev", "next"]

trans_counts = (
    pairs.value_counts()
    .rename("count")
    .reset_index()
)

# Optional: overall popularity for backoff
popularity = (
    df["parent_asin"].value_counts()
    .rename_axis("parent_asin")
    .reset_index(name="pop_count")
)
```

```python
# Python
from collections import defaultdict

def build_next_map(trans_counts_df):
    # next_map[prev][next] = count
    next_map = defaultdict(dict)
    for row in trans_counts_df.itertuples(index=False):
        next_map[row.prev][row.next] = int(row.count)
    return next_map

next_map = build_next_map(trans_counts)

def predict_next_topk(
    context_items,
    next_map,
    popularity_df,
    k=10,
    recency_weights=None,      # list of len(context_items), aligned oldest->newest
    exclude_seen=True,
    min_candidates=200         # if too few candidates from transitions, fill with popularity
):
    if recency_weights is None:
        # default: linearly increasing weights (newest matters most)
        recency_weights = np.linspace(0.2, 1.0, num=len(context_items)).tolist()

    scores = defaultdict(float)

    seen = set(context_items) if exclude_seen else set()

    for prev, w in zip(context_items, recency_weights):
        for nxt, cnt in next_map.get(prev, {}).items():
            if nxt in seen:
                continue
            scores[nxt] += w * cnt

    # Rank by transition score
    ranked = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    preds = [asin for asin, _ in ranked[:k]]

    # Backoff: if we don't have enough candidates, fill with popular items
    if len(ranked) < min_candidates or len(preds) < k:
        pop_list = popularity_df["parent_asin"].tolist()
        for asin in pop_list:
            if exclude_seen and asin in seen:
                continue
            if asin in scores:          # already considered
                continue
            preds.append(asin)
            if len(preds) >= k:
                break

    return preds

top_k = 10
preds = predict_next_topk(
    context_items=context_items,   # oldest -> newest
    next_map=next_map,
    popularity_df=popularity,
    k=top_k,
    exclude_seen=True
)

print("Top-k predictions:", preds)
print("True held-out next item:", true_next_item)
print("Hit@k:", true_next_item in preds)
```
